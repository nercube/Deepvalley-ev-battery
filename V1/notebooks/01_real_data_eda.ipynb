```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv("BMS_data.csv")  # or correct path
df.head()
```

```python
df.shape
df.columns.tolist()
```

```python
df.isna().mean().sort_values(ascending=False)
```

```python
# asset mapping
df["asset_id"] = df["macId"]

# timestamp
df["createdAt"] = pd.to_datetime(df["createdAt"])

# sort strictly
df = df.sort_values(["asset_id", "createdAt"]).reset_index(drop=True)
```

```python
df[["asset_id", "createdAt"]].head()
```

```python
battery_temp_cols = ["temperature[1]", "temperature[2]"]

df["battery_temp_mean"] = df[battery_temp_cols].mean(axis=1)
df["battery_temp_max"]  = df[battery_temp_cols].max(axis=1)
df["battery_temp_min"]  = df[battery_temp_cols].min(axis=1)
```

```python
df["dt_s"] = (
    df.groupby("asset_id")["createdAt"]
      .diff()
      .dt.total_seconds()
)

df.loc[df["dt_s"] <= 0, "dt_s"] = np.nan
```

```python
df["dV"] = df.groupby("asset_id")["pack_voltage"].diff()
df["dV_dt"] = df["dV"] / df["dt_s"]
```

```python
SOC_DROP_THRESHOLD = 5          # %
TIME_GAP_THRESHOLD = 30 * 60    # 30 minutes

df["cycle_id"] = 0

for asset_id, idxs in df.groupby("asset_id").groups.items():
    idxs = list(idxs)

    cycle = 0
    start_soc = df.loc[idxs[0], "soc"]
    last_time = df.loc[idxs[0], "createdAt"]

    for i in idxs:
        current_soc = df.loc[i, "soc"]
        current_time = df.loc[i, "createdAt"]

        soc_drop = start_soc - current_soc
        time_gap = (current_time - last_time).total_seconds()

        if soc_drop >= SOC_DROP_THRESHOLD or time_gap >= TIME_GAP_THRESHOLD:
            cycle += 1
            start_soc = current_soc

        df.at[i, "cycle_id"] = cycle
        last_time = current_time
```

```python
checkpoint_df = (
    df
    .groupby(["asset_id", "cycle_id"], as_index=False)
    .agg(
        V_mean=("pack_voltage", "mean"),
        V_std=("pack_voltage", "std"),
        V_min=("pack_voltage", "min"),
        V_max=("pack_voltage", "max"),
        dV_dt_mean=("dV_dt", "mean"),
        dV_dt_max=("dV_dt", "max"),
        T_mean=("battery_temp_mean", "mean"),
        T_max=("battery_temp_max", "max"),
        T_min=("battery_temp_min", "min"),
        duration_s=("dt_s", "sum"),
    )
)
```

```python
checkpoint_df["V_range"] = checkpoint_df["V_max"] - checkpoint_df["V_min"]
checkpoint_df["T_delta"] = checkpoint_df["T_max"] - checkpoint_df["T_min"]
```

```python
final_checkpoint_df = checkpoint_df[
    [
        "asset_id", "cycle_id",
        "V_mean", "V_std", "V_min", "V_max", "V_range",
        "dV_dt_mean", "dV_dt_max",
        "T_mean", "T_max", "T_delta",
        "duration_s"
    ]
].copy()
```
